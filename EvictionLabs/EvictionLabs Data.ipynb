```python
import re
import pandas as pd
import sys
sys.path.insert(0, '/Users/ewd/ShrinkSmart/')
from ShrinkSmart import ShrinkSmart

df = pd.read_csv('cities.csv')

theCities = list(pd.read_csv("../FullTowns/category-data.csv")['City'].unique())

for city in theCities:
    match = re.match('([^,]+),.*', city)
    if (match is not None):
        cityName = match[1]
    else:
        cityName = ""
    subDF = df[df['name'] == cityName].dropna()
    subDF = subDF.reset_index().drop(axis=1, columns=['GEOID', 'parent-location', 'imputed', 'subbed', 'index', 'pct-multiple', 'pct-other'])
    fileName = "data/" + re.sub("[^\w]+", "_", cityName) + "_IA_evictions.csv"
    subDF.to_csv(fileName)
    
subDF.columns = ['Year', 'City', 'Pop', 'Poverty Rate', '% Renter', 'Median Gross Rent', 'Median Household Income', 'Median Property Value', 'Rent Burden', 'White', 'African American', 'Hispanic', 'American Indian', 'Asian', 'Non-Hispanic', 'Renter Occupied Households', 'Eviction Filings', 'Evictions', 'Eviction Rate', 'Filing Rate']
```

```python
exportDF = df.dropna()
exportDF = exportDF.reset_index().drop(axis=1, columns=['GEOID', 'parent-location', 'imputed', 'subbed', 'index', 'pct-multiple', 'pct-other'])
exportDF.columns = ['Year', 'City', 'Pop', 'Poverty Rate', '% Renter', 'Median Gross Rent', 'Median Household Income', 'Median Property Value', 'Rent Burden', 'White', 'African American', 'Hispanic', 'American Indian', 'Asian', 'Non-Hispanic', 'Renter Occupied Households', 'Eviction Filings', 'Evictions', 'Eviction Rate', 'Filing Rate']

cityBasics = pd.read_csv('../Data/city-basics_iowa.csv')

def getLat(cityName):
    lat = float(cityBasics[cityBasics['City'] == cityName]['Lat'])
    return lat

def getLon(cityName):
    lon = float(cityBasics[cityBasics['City'] == cityName]['Lon'])
    return lon

theData = []
for city in theCities:
    match = re.match('([^,]+),.*', city)
    if (match is not None):
        cityName = match[1]
    else:
        cityName = ""
    sampleSize = len(exportDF[exportDF['City'] == cityName])
    poverty = exportDF[exportDF['City'] == cityName]['Poverty Rate'].sum()/sampleSize
    renter = exportDF[exportDF['City'] == cityName]['% Renter'].sum()/sampleSize
    rent = exportDF[exportDF['City'] == cityName]['Median Gross Rent'].sum()/sampleSize
    income = exportDF[exportDF['City'] == cityName]['Median Household Income'].sum()/sampleSize
    burden = exportDF[exportDF['City'] == cityName]['Rent Burden'].sum()/sampleSize
    eviction = exportDF[exportDF['City'] == cityName]['Eviction Rate'].sum()/sampleSize
    filing = exportDF[exportDF['City'] == cityName]['Filing Rate'].sum()/sampleSize
    row = [cityName, poverty, renter, rent, income, burden, eviction, filing, sampleSize]
    theData.append(row)
    
newDF = pd.DataFrame(theData)
newDF.columns = ["City", "Poverty Rate", "Renter Rate", "Median Rent", "Median Income", "Rent Burden", "Eviction Rate", "Filing Rate", "Samples"]
newDF['City'] = newDF['City'].apply(lambda x: str(x) + ", IA")
newDF['Lat'] = newDF['City'].apply(getLat)
newDF['Lon'] = newDF['City'].apply(getLon)
newDF = newDF.fillna(0.0)
newDF.to_csv("iowa-evictions.csv")
```
