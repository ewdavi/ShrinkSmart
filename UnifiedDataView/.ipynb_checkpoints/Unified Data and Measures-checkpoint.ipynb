```python
import pandas as pd
import geopy.distance
import numpy as np
import sys
sys.path.insert(0, '/Users/ewd/ShrinkSmart/')
from ShrinkSmart import ShrinkSmart

cityDf = pd.read_csv("city-basics_iowa.csv")
cityDf = cityDf.drop(['Unnamed: 0', 'Area (sqmi)', 'Area (sqkm)', 'WikiURL', 'Counties'], axis=1)

childCare = pd.read_csv('iowa-childcare.csv')
childCare = childCare.drop(['Unnamed: 0', 'Lat', 'Lon'], axis=1)

evictions = pd.read_csv('iowa-evictions.csv')
evictions = evictions.drop(['Unnamed: 0', 'Lat', 'Lon'], axis=1)

scalarDf = pd.merge(pd.merge(cityDf, childCare, on="City"), evictions, on="City")

def getDistance(row):
    hLat = row['lat home']
    hLon = row['lon home']
    wLat = row['lat work']
    wLon = row['lon work']
    d = geopy.distance.vincenty((hLat, hLon), (wLat, wLon)).km
    return d

jobsIn = pd.read_csv('iowa-work-flow.csv').drop(['Unnamed: 0'], axis=1)
jobsIn['Distance'] = jobsIn.apply(getDistance, axis=1)

def getPop(city):
    return float(cityDf[cityDf['City'] == city]['Population'])

jobsIn['Population'] = jobsIn['work city'].apply(getPop)

def calcMagnitude(row):
    return row['total jobs'] / row['City Jobs']

cities = list(jobsIn['work city'].unique())
totalJobs = dict()

for city in cities:
    totalJobs[city] = jobsIn[jobsIn['work city'] == city]['total jobs'].sum()
    
totalJobsDF = pd.DataFrame(list(totalJobs.items()), columns=["work city","City Jobs"])

jobsIn = pd.merge(totalJobsDF, jobsIn, on=["work city"])

jobsIn['normalized jobs'] = jobsIn.apply(calcMagnitude, axis=1)
jobsIn['normalized distance'] = jobsIn['Distance'] * jobsIn['normalized jobs']

jobsOut = pd.read_csv('iowa-job-out.csv').drop(["Unnamed: 0"], axis=1)
jobsOut['Distance'] = jobsOut.apply(getDistance, axis=1)
jobsOut['Population'] = jobsOut['home city'].apply(getPop)

cities = list(jobsOut['home city'].unique())
totalJobs = dict()

for city in cities:
    totalJobs[city] = jobsOut[jobsOut['home city'] == city]['total jobs'].sum()
    
totalJobsDF = pd.DataFrame(list(totalJobs.items()), columns=["home city","City Jobs"])

jobsOut = pd.merge(totalJobsDF, jobsOut, on=["home city"])

jobsOut['normalized jobs'] = jobsOut.apply(calcMagnitude, axis=1)
jobsOut['normalized distance'] = jobsOut["Distance"] * jobsOut["normalized jobs"]    

dist50 = dict()
dist90 = dict()
dist95 = dict()
dist99 = dict()
for city in cities:
    q = np.quantile(list(jobsIn[jobsIn['work city'] == city]['normalized distance']), [0.5, 0.9, 0.95, 0.99])
    dist50[city] = q[0]
    dist90[city] = q[1]
    dist95[city] = q[2]
    dist99[city] = q[3]
    
temp = pd.merge(pd.DataFrame(list(dist50.items()), columns=["City", "Q50 In"]), 
                pd.DataFrame(list(dist90.items()), columns=["City", "Q90 In"]), on="City")
temp = pd.merge(temp,
                pd.DataFrame(list(dist95.items()), columns=["City", "Q95 In"]), on="City")
temp = pd.merge(temp,
                pd.DataFrame(list(dist99.items()), columns=["City", "Q99 In"]), on="City")

scalarNew = pd.merge(scalarDf, temp, on="City")

dist50 = dict()
dist90 = dict()
dist95 = dict()
dist99 = dict()
jobMarket = dict()
scaledJM = dict()
for city in cities:
    q = np.quantile(list(jobsOut[jobsOut['home city'] == city]['normalized distance']), [0.5, 0.9, 0.95, 0.99])
    dist50[city] = q[0]
    dist90[city] = q[1]
    dist95[city] = q[2]
    dist99[city] = q[3]
    jobMarket[city] = jobsIn[jobsIn["work city"] == city]["total jobs"].sum()
    
temp = pd.merge(pd.DataFrame(list(dist50.items()), columns=["City", "Q50 Out"]), 
                pd.DataFrame(list(dist90.items()), columns=["City", "Q90 Out"]), on="City")
temp = pd.merge(temp,
                pd.DataFrame(list(dist95.items()), columns=["City", "Q95 Out"]), on="City")
temp = pd.merge(temp,
                pd.DataFrame(list(dist99.items()), columns=["City", "Q99 Out"]), on="City")
temp = pd.merge(temp,
                pd.DataFrame(list(jobMarket.items()), columns=["City", "Job Market"]), on="City")

scalarNew = pd.merge(scalarNew, temp, on="City")
scalarNew["Scaled Job Market"] = scalarNew["Job Market"]/scalarNew["Population"]

propLocalsWorking = jobsOut[jobsOut["home city"] == jobsOut["work city"]][['home city', 'normalized jobs']]
propWorkersLocal = jobsIn[jobsIn["home city"] == jobsIn["work city"]][['home city', 'normalized jobs']]
propLocalsWorking.columns = ['City', 'Prop Locals Working']
propWorkersLocal.columns = ['City', 'Prop Workers Local']
temp = pd.merge(propLocalsWorking, propWorkersLocal, on="City")
scalarNew = pd.merge(temp, scalarNew, on="City")
scalarNew = scalarNew.rename(columns={"Capacity" : "Child Care Capacity"})
scalarNew['Scaled Child Care'] = scalarNew['Child Care Capacity'] / scalarNew['Population']
```

```python
import os
import re

servicesDF = pd.DataFrame(columns=["City", "destination", "km distance", "rating", "num reviews", "label"])
servicesDF.reset_index()

def getCity(address):
    m = re.match(".*?,?([\s\w]+,\s?[A-Z][A-Z]\s?).*", address)
    if m is None:
        print(address)
    else:
        m = m[1]
    return(m)

cityData = dict()
yelpFiles = os.listdir("/Users/ewd/ShrinkSmart/Yelp/")

theCities = list(pd.read_csv('../FullTowns/category-data.csv')['City'].unique())

for city in theCities:
    fileindicator = str(re.sub(", ", "_", city))
    cityData[city] = [x for x in yelpFiles if fileindicator in x]
    for file in cityData[city]:
        file = "../Yelp/" + file
        df = pd.read_csv(file)
        df = df[~df["permanently closed?"]][["origination", "city", "address", "km distance", "rating", "num reviews", "category"]]
        df = df.rename(columns={"category":"label", "city":"destination", "origination":"City"})
        df["destination"] = df["address"].apply(getCity)
        #df["destination"] = df["destination"].apply(lambda x: x + ", IA")
        servicesDF = pd.concat([servicesDF, df])

servicesDF = servicesDF.reset_index().drop(["index"], axis=1).drop(["address"], axis=1)
servicesDF.to_csv("connection-metrics-services.csv")
scalarNew.to_csv("scalar-metrics.csv")
jobsIn.to_csv("jobsIn-metrics.csv")
jobsOut.to_csv("jobsOut-metrics.csv")
```

```python
scalarFeatures = pd.merge(scalarNew, pd.read_csv("services-scalars.csv").drop(columns=["Unnamed: 0"], axis=1), on="City")
scalarFeatures["category"] = scalarFeatures["City"].apply(lambda x: ShrinkSmart.getCityClass(x))
categories = list(set(ShrinkSmart.categories))
categoryID = [1, 0, 1, 0] #list(range(0, len(categories)))
categoryDict = dict(zip(categories, categoryID))
idDict = dict(zip(categoryID, categories))
scalarFeatures["category id"] = scalarFeatures["category"].apply(lambda x: categoryDict[x])


data = dict()

features = scalarFeatures.drop(["Lat", "Lon", "Population"], axis=1)

dataDF = scalarFeatures.drop(columns=["Lat", "Lon", "Samples", "Population", "City", "category", "category id"])

data["feature names"] = list(dataDF.columns)
data["data"] = np.array(dataDF)
data["target"] = np.array(features["category id"])
scalarFeatures.to_csv("scalar-features-full.csv")
categoryDict
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.datasets import make_classification

mse = 0.0
runs = 1000.0

results = pd.DataFrame()

results["City"] = features["City"]
results["Metric"] = 0
results["Category"] = features["category"]

for idx in range(0,int(runs)):
    
    training = dict()
    testing = dict()
    training["data"], testing["data"], training["target"], testing["target"] = train_test_split(data["data"], data["target"], test_size=0.2)

    #clf = RandomForestClassifier(n_estimators=500, n_jobs=-1)
    #clf = KNeighborsClassifier(n_neighbors=5)
    clf = svm.SVC(gamma='scale', decision_function_shape='ovr')
    clf.fit(training["data"], training["target"])
    #for name, importance in zip(data["feature names"], clf.feature_importances_):
    #    print(name, "=", importance)
    testing = data
    
    predictions_rf = clf.predict(testing["data"])
    mse += mean_squared_error(predictions_rf, testing["target"])
    outcomes = pd.DataFrame({"Truth": testing["target"], "Prediction": predictions_rf})
    outcomes["Truth"] = outcomes["Truth"].apply(lambda x: idDict[x])
    outcomes["Prediction"] = outcomes["Prediction"].apply(lambda x: idDict[x])
    outcomes["Success"] = outcomes["Truth"] == outcomes["Prediction"]
    results["Metric"] += outcomes["Truth"] == outcomes["Prediction"]
    outcomes
    
results["Metric"] /= runs
mse /= runs
print(mse)
results.sort_values(by=["Metric"], ascending=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
%matplotlib inline

mse = 0.0
runs = 10.0

results = pd.DataFrame()

results["City"] = features["City"]
results["Metric"] = 0
results["Category"] = features["category"]

clf = SVC(kernel="linear")
#clf = RandomForestClassifier(n_estimators=500, n_jobs=-1)
rfecv = RFECV(estimator=clf, step=1, cv=StratifiedKFold(2), scoring='accuracy')
rfecv.fit(data["data"], data["target"])
    
print("Optimal number of features : %d" % rfecv.n_features_)

# Plot number of features VS. cross-validation scores
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```
