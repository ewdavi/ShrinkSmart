```python
import sys
sys.path.insert(0, '/Users/ewd/ShrinkSmart/')
from ShrinkSmart import ShrinkSmart
import pandas as pd
from math import *
from tqdm import tnrange, tqdm_notebook

def distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = (sin(dlat/2))**2 + cos(lat1) * cos(lat2) * (sin(dlon/2))**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return(distance)

blockDf = pd.read_csv("block-city-mapping.csv")
cityDf = pd.read_csv("city-basics_iowa.csv")
cityData = cityDf

blockDf["Census Block ID"] = blockDf["Census Block ID"].apply(lambda x: str(x))

cityData = cityData[cityData['City'].isin(ShrinkSmart.cities)]
```

```python
def getClosestStudyCity(row):
    shortD = 100000
    shortC = "ERROR"
    cLat = 0.0
    cLon = 0.0
    for city in ShrinkSmart.cities:
        bLat = row["Lat"]
        bLon = row["Lon"]
        (cLat, cLon) = ShrinkSmart.getLatLong(city)
        d = distance(bLat, bLon, cLat, cLon)
        if (d < shortD):
            shortD = d
            shortC = city
    return(shortC)

blockDf["Closest Study City"] = blockDf.apply(getClosestStudyCity, axis=1)
blockDf
```

```python
def findStudyDistance(row):
    bLat = row["Lat"]
    bLon = row["Lon"]
    city = row["Closest Study City"]
    (cLat, cLon) = ShrinkSmart.getLatLong(city)
    d = distance(bLat, bLon, cLat, cLon)
    return(d)

blockDf["Study City Distance"] = blockDf.apply(findStudyDistance, axis=1)
```

```python
maxDist = (blockDf[blockDf["Closest Study City"] == blockDf["City"]]["Study City Distance"]).max()

def fixCensusBlockBias(row):
    city = row["City"]
    if (row["Closest Study City"] != row["City"]) & (row["Study City Distance"] < maxDist):
        city = row["Closest Study City"]
    return(city)

blockDf["City"] = blockDf.apply(fixCensusBlockBias, axis=1)
```

```python
blockDf = blockDf.drop(["Unnamed: 0", "Closest Study City", "Study City Distance"], axis=1)
blockDf.to_csv("new-block-city-mappings.csv")
```
